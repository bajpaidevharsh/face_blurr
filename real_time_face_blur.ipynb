```python
import numpy
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

**in the path of image give path of image in which you want ot blur face**

```python
img=cv2.imread('#path of image')
```

```python
cv2_imshow(img)
cv2.waitKey(0)
```

```python
detector = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
face=detector.detectMultiScale(img,1.1,4)
```

```python
for (x, y, w, h) in face:
        frame2 = img[y:y+h, x:x+w]
        blurred_region = cv2.GaussianBlur(frame2, (35, 35), 1000)
        img[y:y+h, x:x+w] = blurred_region
        cv2_imshow(img)
        cv2.waitKey(0)
```

**for real time blurring**

**for face blurring in video give path of video or for using webcam use 0 for default cam of device**

```python
cap=cv2.VideoCapture("#path to the video")
while True:
    success,frame=cap.read()
    if success:
        face=detector.detectMultiScale(frame,1.1,4)
        for x,y,w,h in face:
            #cv2.rectangle(frame,(x,y),(x+h,y+w),(0,0,255),4)
            frame2=frame[y:y+h,x:x+w]
            blurred_region = cv2.GaussianBlur(frame2, (35, 35), 1000)
            frame[y:y+h,x:x+w] = blurred_region
        cv2.imshow('output2',frame)
        if cv2.waitKey(1) & 0xFF == ord('c'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
